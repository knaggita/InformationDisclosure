```python
import sys
sys.path.append('../')

import numpy as np
import pandas as pd 

from scripts.analytics import *
from scripts.algorithms import *
```

```python
name = "studentmath"
namex = "Student-mat "
d = 30

###knn
graphs = np.load("../graphs/"+name+"_knn_random.npy", allow_pickle=True)
###threshold
graphsx = np.load("../graphs/"+name+"_thresh_random.npy", allow_pickle=True)
```

# Compute greedy and random algorithms results# Compute greedy and random algorithms results

## Nearest neighbor

```python
summary_df = pd.DataFrame()
greedyalgo_df = pd.DataFrame()
randomalgo_df = pd.DataFrame()

for lz in range(len(graphs)):

    r_edges_random  = graphs[lz]["edges"]
    r_labels_random = graphs[lz]["labels"]

    for Kz in [1, 2, 3, 4, 5]:
        
        df_greedy, df_random, dfz = heuristic_greedy_random(edgesx=r_edges_random, 
                                                            labelsx=r_labels_random, budgetx=Kz)

        dfz["kmax"] =  graphs[lz]["k_max"]
        dfz["n"] = graphs[lz]['n']
        dfz["m"] =  graphs[lz]['m']
        dfz["dataset"] = name+" ("+ str(d) + ")"
        dfz["graphid"] = lz
        
        df_greedy["graphid"] = [lz]*(len(df_greedy))
        df_random["graphid"] = [lz]*(len(df_random))
        
        summary_df = pd.concat([summary_df, dfz], ignore_index=True)
        greedyalgo_df = pd.concat([greedyalgo_df, df_greedy], ignore_index=True)
        randomalgo_df = pd.concat([randomalgo_df, df_random], ignore_index=True)
```

## Thresholding

```python
summary_dfx = pd.DataFrame()
greedyalgo_dfx = pd.DataFrame()
randomalgo_dfx = pd.DataFrame()

for lx in range(len(graphsx)):

    r_edges_randomx  = graphsx[lx]["edges"]
    r_labels_randomx  = graphsx[lx]["labels"]

    for Kx in [1, 2, 3, 4, 5]:
        
        df_greedy, df_random, dfx = heuristic_greedy_random(edgesx=r_edges_randomx, 
                                                            labelsx=r_labels_randomx, budgetx=Kx)

        dfx["r"] =  graphsx[lx]["threshold"]
        dfx["n"] = graphsx[lx]['n']
        dfx["m"] =  graphsx[lx]['m']
        dfx["dataset"] = name+" ("+ str(d) + ")"
        dfx["graphid"] = lx
        
        df_greedy["graphid"] = [lx]*(len(df_greedy))
        df_random["graphid"] = [lx]*(len(df_random))
        
        summary_dfx = pd.concat([summary_dfx, dfx], ignore_index=True)
        greedyalgo_dfx = pd.concat([greedyalgo_dfx, df_greedy], ignore_index=True)
        randomalgo_dfx = pd.concat([randomalgo_dfx, df_random], ignore_index=True)
```
